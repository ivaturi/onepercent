## Chapter 4: Statistics <a name="statistics"></a>

```python
from lib import stats
from lib import matrix as mat
from lib import vector as vec
```

```python
v = vec.create_random(20)

print(f"            Vector   : {v}")
print(f"            Mean     : {stats.mean(v)}")
print(f"            Median   : {stats.median(v)}")
print(f"            Range    : {stats.range(v)}")
print(f"            Min      : {stats.min(v)}")
print(f"            Max      : {stats.max(v)}")
print(f"            Variance : {stats.variance(v):.2f}")
print(f"            Std. Dev : {stats.std_dev(v):.2f}")
print(f" Interquantile range : {stats.interquantile_range(v)}")
```

```python
v1 = v
v2 = vec.create_random(len(v1))

print(f" Vector2 : {v2}")
print(f" Covariance: {stats.covariance(v1, v2):.2f}")
print(f" Correlation: {stats.correlation(v1, v2):.2f}")
```

## Chapter 5: Probability <a name="probability"></a>

```python
from lib import prob
from matplotlib import pyplot as plt
```

```python
# uniform pdf
x = [i*0.1 for i in range(-10,20)]
y = [prob.pdf_uniform(xi) for xi in x]

plt.plot(x,y)
plt.title("Uniform probability density")
plt.show()
```

```python
# uniform cdf
plt.plot(x, [prob.cdf_uniform(xi) for xi in x])
plt.title("Uniform cumulative density")
plt.show()
```

```python
# normal pdf
mu = 0
x = range(-50,50)
plt.plot(x, [prob.pdf_normal(xi, mu, 10) for xi in x], "--", label="sigma=10")
plt.plot(x, [prob.pdf_normal(xi, mu, 5) for xi in x], ".-", label="sigma=5")
plt.legend()
plt.title("Normal probability density")
plt.show()
```
