```python
# an elemental neural network - this one takes the input, applies a scaling factor to it, and returns the result as a prediction
def nn1(input, weight = 0.5):
    # scale input by 0.5 and return prediction
    pred = input * weight
    return pred

w = [10, -10, 20, 5000]
[nn1(wi) for wi in w]
```

```python
# this one takes multiple inputs and computes a normalized weighted sum as its prediction
# weights are automagically guessed by a random number generator
import random
def nn_ws(*inputs):
    weights = [random.random() for i in range(len(inputs))]
    ws = [i*wi for i,wi in zip(weights, inputs)]
    pred = sum(ws)/sum(inputs)
    return pred

nn_ws(10, 20, 13, 12, 15, 20)
```
